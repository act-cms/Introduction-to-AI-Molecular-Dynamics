# AI-Assisted MD Workflows

AI tools and large language models can help with molecular dynamics workflows, but they must be used with verification. This notebook introduces practical ways to use AI for planning, scripting, analysis, and review.

## 1. Useful AI Roles in MD

AI assistants are often useful for:

- Drafting first-pass LAMMPS or Python analysis scripts.
- Explaining unfamiliar simulation commands.
- Creating checklists for validation.
- Summarizing log files or error messages.
- Helping plan parameter sweeps.
- Turning results into plots, tables, or teaching materials.

They are not a substitute for testing, physical reasoning, or checking source documentation.

```python
from pathlib import Path

for dirname in ["prompts", "lammps", "analysis", "figures"]:
    Path(dirname).mkdir(exist_ok=True)

prompt = """You are helping me create a LAMMPS input script for a beginner MD lesson.

Goal: simulate a Lennard-Jones FCC solid in reduced units.
Requirements:
- Use units lj and atom_style atomic.
- Create the atoms directly in LAMMPS.
- Use pair_style lj/cut.
- Run a short NVE simulation.
- Write thermo output and a dump file for OVITO.
- Explain each command briefly after the script.

Before finalizing, list assumptions and possible mistakes to check.
"""

Path("prompts/lammps_lj_prompt.txt").write_text(prompt)
print(prompt)
```

## 2. AI Output Audit Checklist

When an AI tool generates MD code, review it before running:

- Are the units consistent?
- Is the atom style appropriate?
- Does the potential match the material and atom types?
- Are boundary conditions physically meaningful?
- Is the time step reasonable?
- Is the thermostat or barostat appropriate for the ensemble?
- Are outputs sufficient for validation and visualization?
- Does the script actually run on a small test case?

```python
audit_items = [
    "units consistent",
    "atom style appropriate",
    "potential matches material",
    "boundary conditions justified",
    "time step reasonable",
    "ensemble commands correct",
    "thermo and dump outputs present",
    "small test run completed",
]

for item in audit_items:
    print(f"[ ] {item}")
```

## 3. Parameter Sweep Planning

AI tools can help plan systematic simulation campaigns. The code below creates a simple parameter table for a temperature sweep. In a real project, each row could become one simulation directory.

```python
import pandas as pd

temperatures = [100, 200, 300, 400, 500]
runs = []

for temp in temperatures:
    runs.append({
        "run_id": f"T{temp}",
        "temperature_K": temp,
        "ensemble": "NVT",
        "equilibration_steps": 20000,
        "production_steps": 50000,
        "status": "planned",
    })

sweep = pd.DataFrame(runs)
sweep.to_csv("analysis/temperature_sweep_plan.csv", index=False)
sweep
```

## 4. Responsible Use

AI-assisted MD work should remain reproducible and inspectable. Keep prompts, generated code, corrections, run logs, and validation plots together. Record which suggestions were accepted, rejected, or modified.
