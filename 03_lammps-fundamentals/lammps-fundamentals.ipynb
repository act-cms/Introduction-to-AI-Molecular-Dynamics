# LAMMPS Fundamentals

LAMMPS is a widely used molecular dynamics code for simulating atoms, molecules, and materials. This notebook introduces the structure of a LAMMPS input script and creates a minimal Lennard-Jones example that can be run outside the notebook with LAMMPS installed.

## 1. Anatomy of a LAMMPS Input Script

Most LAMMPS scripts contain the same broad sections:

1. Simulation settings: units, dimensions, boundary conditions, atom style.
2. System construction or data import.
3. Force field definition.
4. Thermodynamic output and trajectory dumps.
5. Minimization, equilibration, or production dynamics.

````python
from pathlib import Path

lammps_dir = Path("lammps")
lammps_dir.mkdir(exist_ok=True)

lj_script = """# Minimal Lennard-Jones solid example
units           lj
dimension       3
boundary        p p p
atom_style      atomic

lattice         fcc 0.8442
region          box block 0 6 0 6 0 6
create_box      1 box
create_atoms    1 box
mass            1 1.0

pair_style      lj/cut 2.5
pair_coeff      1 1 1.0 1.0 2.5

velocity        all create 0.8 12345 mom yes rot yes dist gaussian
neighbor        0.3 bin
neigh_modify    every 1 delay 0 check yes

thermo          100
thermo_style    custom step temp pe ke etotal press
dump            traj all atom 100 dump.lj

fix             int all nve
run             2000
"""

script_path = lammps_dir / "in.lj_solid"
script_path.write_text(lj_script)
print(script_path.resolve())
````

## 2. Run Command

If LAMMPS is installed and available on your command line, run this from the lesson folder:

```bash
lmp -in lammps/in.lj_solid
```

Some installations use a different executable name, such as `lmp_serial`, `lmp_mpi`, or `lammps`.

````python
print(lj_script)
````

## 3. Key Commands to Understand

- `units lj`: uses reduced Lennard-Jones units.
- `lattice fcc 0.8442`: defines an FCC lattice at a chosen reduced density.
- `create_box` and `create_atoms`: build the simulation cell and atoms.
- `pair_style lj/cut`: selects the interaction model.
- `fix nve`: integrates equations of motion at constant energy.
- `dump`: writes trajectory frames for visualization in OVITO.
