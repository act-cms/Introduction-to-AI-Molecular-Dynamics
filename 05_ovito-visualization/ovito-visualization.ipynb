# OVITO Visualization

OVITO is a visualization and analysis tool for atomistic simulation data. This notebook introduces a workflow for loading LAMMPS output, applying common modifiers, and preparing figures or movies.

## 1. Files OVITO Commonly Opens

OVITO can read many MD file formats, including:

- LAMMPS dump files.
- LAMMPS data files.
- XYZ files.
- CFG files from Atomsk.
- POSCAR/CONTCAR files from VASP workflows.

For this course, the most common workflow is: build or simulate a structure, write a LAMMPS dump file, then inspect it in OVITO.

```python
from pathlib import Path

ovito_dir = Path("ovito")
ovito_dir.mkdir(exist_ok=True)

checklist = """OVITO visualization checklist

1. File > Load File and select a LAMMPS dump or data file.
2. Confirm the simulation cell is correct.
3. Set particle coloring by type, potential energy, or displacement.
4. Add modifiers such as Coordination Analysis, Common Neighbor Analysis, or Atomic Strain.
5. Use Slice or Select Type to simplify crowded views.
6. Adjust camera, lighting, and background.
7. Export a snapshot or animation.
"""

(ovito_dir / "visualization_checklist.txt").write_text(checklist)
print(checklist)
```

## 2. Useful OVITO Modifiers

| Modifier | Use |
|---|---|
| Coordination Analysis | Count neighbors and identify local density changes |
| Common Neighbor Analysis | Identify FCC, HCP, BCC, and disordered environments |
| Atomic Strain | Measure local deformation relative to a reference |
| Dislocation Analysis | Extract dislocation lines in crystalline materials |
| Voronoi Analysis | Compute local atomic volumes and coordination |
| Cluster Analysis | Detect connected groups of selected atoms |

```python
ovito_python_example = """# Optional OVITO Python example
# Run with ovitos or inside an OVITO Python environment.

from ovito.io import import_file, export_file
from ovito.modifiers import CommonNeighborAnalysisModifier

pipeline = import_file('data/dump.lammpstrj')
pipeline.modifiers.append(CommonNeighborAnalysisModifier())
data = pipeline.compute()

export_file(pipeline, 'figures/structure_with_cna.png', 'png')
"""

(ovito_dir / "ovito_python_example.py").write_text(ovito_python_example)
print(ovito_python_example)
```

## 3. Figure Design Tips

- Show the atoms or defect clearly before adding annotations.
- Use color to encode one important variable at a time.
- Crop or slice dense structures when the full cell is visually confusing.
- Keep camera angles consistent when comparing before and after states.
- Export high-resolution images for reports and slides.
